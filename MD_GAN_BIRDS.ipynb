```python
with open('train_test_split.txt') as train_test:
    lines = train_test.read().splitlines()
with open('images.txt') as image_add:
    images = image_add.read().splitlines()
```

```python
import os
main_dst = '/content/DM-GAN/data/birds/CUB_200_2011'
test_dir = main_dst + '/test/'
train_dir = main_dst + '/train/'
os.mkdir(test_dir)
os.mkdir(train_dir)
```

```python
for image_num, image in enumerate(images):
  imagefile=image.split(' ')[1]
  try:
    dst = train_dir + imagefile.split('/')[0] 
    os.mkdir(dst)
    dst = test_dir + imagefile.split('/')[0] 
    os.mkdir(dst)
  except:
    pass
```

```python
import shutil
image_dir = main_dst + '/images/'
for i, image in enumerate(images):
  imagefile=image.split(' ')[1]
  is_train=lines[i].split(' ')[1]
  if is_train =='0':
    dst = train_dir + imagefile.split('/')[0]  
    shutil.copy(image_dir + imagefile, dst)
  else:
    dst = test_dir + imagefile.split('/')[0]  
    shutil.copy(image_dir + imagefile, dst)
```
